```python
%pylab inline
from __future__ import division

import os
import sys

from linac import Linac, Simulation
```

```python
mm = 1

x_dim = 101
y_dim = 101
z_dim = 100

x_min = -75.75 * mm
y_min = -75.75 * mm
z_min = -50 * mm

x_max = 75.75 * mm
y_max = 75.75 * mm
z_max = 0 * mm

x_res = (x_max - x_min) / (x_dim)
y_res = (y_max - y_min) / (y_dim)
z_res = (z_max - z_min) / (z_dim)

x = arange(x_min,x_max,x_res) + x_res/2
y = arange(y_min,y_max,y_res) + y_res/2
z = -(arange(z_min,z_max,z_res) + z_res/2)
```

```python
sim_name = "nonreal"
runid = 22644

data_directory = "histograms"
run_name = "basic"
```

```python
energy = load("%s/energy_%s_%s_%s.npy" % (data_directory, sim_name, run_name, runid))
```

```python
plot(z,energy[50,50,:])
grid(True)
```

```python
depth_sum = sum(sum(energy,axis=0),axis=0)

plot(z,depth_sum)
grid(True)
```

```python
contour(x,y,energy[:,:,amax(depth_sum)])
axis('equal')
grid(True)
```

```python
sum_profile = sum(energy, axis=2)

contourf(x,y,sum_profile,50)
axis('equal')
grid(True)
```

```python
plot(x,energy[50,:,amax(depth_sum)])
grid(True)
```

```python
plot(x,sum(energy,axis=1)[:,amax(depth_sum)])
grid(True)
```
